```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "000001.png"
```

# PIL打开保存8bitmask(彩色和黑白都正确)

```python
image1 = Image.open(image_path)
image1
```

```python
# 直接保存还能有颜色
image1.save("000001-1.png")
```

```python
np.unique(image1)
```

```python
# 转换为numpy后再转换为Image就没有颜色了
Image.fromarray(np.array(image1), mode="L").save("000001-2.png")
```

```python
np.unique(Image.open("000001-2.png"))
```

# opencv打开保存mask

## 读取8bit彩色mask(错误)

```python
image2 = cv2.imread(image_path, flags=0)
np.unique(image2)  # 标签错误
```

```python
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
```

```python
image3 = cv2.imread("000001-1.png", flags=0)
np.unique(image3)  # 标签错误
```

```python
plt.imshow(cv2.cvtColor(image3, cv2.COLOR_BGR2RGB))
```

## 读取8bit黑白mask(正确)

```python
image4 = cv2.imread("000001-2.png", flags=0)
np.unique(image4)  # 标签正确
```

```python
plt.imshow(cv2.cvtColor(image4, cv2.COLOR_BGR2RGB))
```
